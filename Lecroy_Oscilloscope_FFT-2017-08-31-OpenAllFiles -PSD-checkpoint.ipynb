```python
# This is for HF TC parameter caluclation
__Author__='Sung Chung'
import numpy as np
import matplotlib.pyplot as plt  
import pandas as pd
import math
from scipy.fftpack import fft
import glob
import ntpath
from scipy import signal


#%matplotlib auto
%matplotlib inline


def path_leaf(path):
    head, tail = ntpath.split(path)
    longName= (tail or ntpath.basename(head))
    return longName.split('.')[0]

directory= "C:\Users\chung_su\Box Sync\__temp"




csvNames=glob.glob(directory + '//'+"*.csv")   # csv or txt

filenames=[path_leaf(path) for path in csvNames]
Nfile= len(filenames)

for filename in filenames:

    ####################
    filepath=directory+"\\"+filename+".csv"
    #____________________________________
    DataFile=filename + "_DataTable.csv"
    PSDFile=filename + "_PSD.csv"
    sumName=filename + "_Summary.csv"
    figureName=filename+'_pic.png'

    sumf=directory+"\\"+sumName
    PSDf=directory+"\\"+PSDFile
    dataf=directory+"\\"+DataFile
    picf=directory+'\\'+figureName

    df= pd.read_csv(filepath,skiprows=4)

    y=np.array(df['Ampl'])
    N=np.float(y.size)     #N=40002
    Fs=20e+9       #sampling rate  20000000000.0
    T=(1.0/Fs)
    
    
    f, Pxx_den = signal.periodogram(y, Fs)
    psdf=pd.DataFrame({'freq_Hz':f,
                   'Amp':Pxx_den})
    psdf.to_csv(PSDf)

    yf = fft(y)
    xf = np.linspace(0.0, 1.0/(2.0*T), N/2)         # up to 10GHz (half of 20GHz sampling rate)
    xf=xf/(1*1e6)        #   MHz

    yhalf=int (N/2)                                #to match xf length   
    xL= int (N*0.0002)
    yf2=abs(yf[0:yhalf])

    sumdf=pd.DataFrame({'freq_MHz':xf,
                       'Amp':yf2,})

    sumdf.to_csv(dataf)

    sumdf.head()

    plt.figure(figsize=(10, 3), dpi= 80, facecolor='y', edgecolor='k')
    plt.title('filename')

    plt.subplot(1,3,1)
    xfG=xf/(1*1e3)           # to make GHz from MHz
    plt.plot(xfG,yf2)
    plt.yscale('log')
    plt.xlabel('GHz')
    plt.title(filename)


    plt.subplot(1,3,2)
    plt.semilogy(f[1:]/1000000, Pxx_den[1:])
    plt.xlabel('MHz')

    plt.subplot(1,3,3)
    plt.plot(xf,yf2)
    plt.xlim([xf[0], xf[xL]])
    plt.ylim([0, 140])
    plt.xlabel('MHz')

    #plt.show()
    plt.savefig(picf)
```
